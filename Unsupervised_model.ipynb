```python
import pandas as pd
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
import joblib  # for saving the model later
```

```python
df = pd.read_csv("processed_logs.csv")  # update the filename as needed
df.head()  # peek at the first few rows
```

```python
X = df.to_numpy()
```

```python
model = IsolationForest(
    n_estimators=100,       # number of trees
    contamination=0.05,     # estimated anomaly proportion
    random_state=42         # ensures reproducibility
)
model.fit(X)
```

```python
predictions = model.predict(X)  # returns 1 for normal, -1 for anomaly
scores = model.decision_function(X)  # lower score = more suspicious
```

```python
df["prediction"] = predictions
df["anomaly_score"] = scores
df.head()
```

```python
df[df['prediction'] == -1]
```

```python
joblib.dump(model, "isolation_forest_model.pkl")
```
